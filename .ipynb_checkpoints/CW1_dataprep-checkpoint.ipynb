```python
import pandas as pd
with open('./data/train.csv') as csv_file: 
    df = pd.read_csv(csv_file)
list(df) 
```

```python
# Prepare test set
test = df.bidid.copy()
for variable in ['weekday', 'hour', 'useragent', 'region', 'adexchange', 'slotvisibility', 'slotformat', 'advertiser']:
    tmp = pd.get_dummies(df[variable], prefix=variable)
    test = pd.concat([test, tmp], axis=1)

tmp = pd.Series([df.slotheight.apply(str) + '_' + df.slotwidth.apply(str)])
tmp = pd.get_dummies(tmp[0], prefix='slotsize')
test = pd.concat([test, tmp], axis=1)
test.rename(columns={0:'slotsize'}, inplace=True)

test = pd.concat([test, df.slotprice], axis=1)


# Prepare training set
train = df.click.copy()
for variable in ['weekday', 'hour', 'useragent', 'region', 'adexchange', 'slotvisibility', 'slotformat', 'advertiser']:
    tmp = pd.get_dummies(df[variable], prefix=variable)
    train = pd.concat([train, tmp], axis=1)

tmp = pd.Series([df.slotheight.apply(str) + '_' + df.slotwidth.apply(str)])
tmp = pd.get_dummies(tmp[0], prefix='slotsize')
train = pd.concat([train, tmp], axis=1)
train.rename(columns={0:'slotsize'}, inplace=True)

train = pd.concat([train, df.slotprice, df.bidprice, df.payprice], axis=1)
#print('Training set variables:\n', list(train))

# Prepare validation set
validation = df.click.copy()
for variable in ['weekday', 'hour', 'useragent', 'region', 'adexchange', 'slotvisibility', 'slotformat', 'advertiser']:
    tmp = pd.get_dummies(df[variable], prefix=variable)
    validation = pd.concat([validation, tmp], axis=1)

tmp = pd.Series([df.slotheight.apply(str) + '_' + df.slotwidth.apply(str)])
tmp = pd.get_dummies(tmp[0], prefix='slotsize')
validation = pd.concat([validation, tmp], axis=1)
validation.rename(columns={0:'slotsize'}, inplace=True)

validation = pd.concat([validation, df.slotprice, df.bidprice, df.payprice], axis=1)
#print('\nValidation set variables:\n', list(train))
list(validation)
```

 ### Evaluate Constant Bid Strategy on Training set :
    

```python
max_bid = 300
budget = 6250 # budget in fen 

def eval_bid(bid,dataset = df):
    # max number of wins at this bid level given budget constraint
    possible_wins = int(budget/(bid/1000))

    # Total number of wins at this bid level
    df['wins'] = df.payprice < bid
    total_wins = df['wins'].sum()
    print(total_wins)

    # The total number of clicks if no budget constraint with this bid level
    df['won_clicks'] = df['wins'] * df['click']
    max_clicks = df['won_clicks'].sum()
    
    # Expected number of clicks and ctr at this bid level
    exp_clicks = max_clicks*min(possible_wins/total_wins,1)
    ctr = exp_clicks/min(possible_wins,total_wins)
    
    # Remaining budget:
    rem_budget = budget - min(possible_wins,total_wins)*bid/1000

    return bid, round(exp_clicks,10), round(ctr*100,30), rem_budget

    
def optimal_bid(dataset=df):
    best_numclick = -1
    best_bid = -1
    
    for bid in range(1,max_bid):
        price, exp_clicks, ctr, rem_budget = eval_bid(bid)
        
        if exp_clicks > best_numclick:
            best_numclick = exp_clicks
            best_bid = price
            best_ctr = ctr
            remaining_cash = rem_budget
            
    return best_bid,best_numclick, best_ctr, remaining_cash

####################

# print(" Best bid: {} \n Expected number of clicks: {} \n CTR: {} \n Remaining cash: {} ".format(*optimal_bid()))
    
```

```python
# double checking results vidually 
for bid in range(1,max_bid):
    print(eval_bid(bid))
```
